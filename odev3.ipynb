```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega_array = np.arange(np.pi*-2,np.pi*2, 0.001)

def response(omega_array, a):
    result = []
    for omega in omega_array:
        result.append(1/ (1-(a*(np.exp(-1j*omega)))))
    return np.array(result)

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7,ax8)) = plt.subplots(4, 2, figsize=(16,12))

result = response(omega_array, 0.2)
ax1.plot(omega_array, abs(result))
ax2.plot(omega_array, np.angle(result))

result = response(omega_array, 0.5)
ax3.plot(omega_array, abs(result),)
ax4.plot(omega_array, np.angle(result))

result = response(omega_array, 0.8)
ax5.plot(omega_array, abs(result))
ax6.plot(omega_array, np.angle(result))

result = response(omega_array, 0.99)
ax7.plot(omega_array, abs(result))
ax8.plot(omega_array, np.angle(result))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega_array = np.arange(np.pi*-2,np.pi*2, 0.001)

def response(omega_array, n):
    result = []
    for omega in omega_array:
        result.append((np.exp(-1j*omega*n)-1) / (np.exp(-1j*omega)-1))
    return np.array(result)

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2, figsize=(16,12))

result = response(omega_array, 6)
ax1.plot(omega_array, abs(result))
ax2.plot(omega_array, np.angle(result))

result = response(omega_array, 9)
ax3.plot(omega_array, abs(result),)
ax4.plot(omega_array, np.angle(result))

result = response(omega_array, 15)
ax5.plot(omega_array, abs(result))
ax6.plot(omega_array, np.angle(result))
```
